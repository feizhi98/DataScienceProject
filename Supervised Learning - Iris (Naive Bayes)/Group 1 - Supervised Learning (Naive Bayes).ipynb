Feature selection is the process of reducing the number of input variables when developing a predictive model. It is desirable to reduce the number of input variables to both reduce the computational cost of modeling and, in some cases, to improve the performance of the model.

#### Method for feature selection: Filtering Method (Anova Method)

The ANOVA F-value approach calculates the degree of linearity between the predictor and the input feature. A high F-value implies good linearity, while a low F-value suggests low linearity. The biggest limitation of utilising ANOVA F-values is that they only capture linear connections between input and output features. In other words, F-value cannot detect any non-linear connections.

#### Gaussian Naive Bayes

The Naive Bayes classification algorithm is used to solve binary (two-class) and multi-class classification issues. When stated using binary or categorical input values, the technique is simplest to grasp.

The method is known as naive Bayes or stupid Bayes because the probabilities for each hypothesis are simplified to make their calculation tractable. Instead of attempting to determine the values of each characteristic P(d1, d2, d3|h), they are considered to be conditionally independent given the goal value and calculated as P(d1|h) * P(d2|H), and so on.

This is a very strong assumption that is highly unlikely to be true in real-world data, namely that the qualities do not interact. Nonetheless, the technique outperforms on data where this assumption is violated.

## Topic 6 : Optimization

Optimization is the process where we train the model iteratively that results in a maximum and minimum function evaluation. It is one of the most important phenomena in Machine Learning to get better results.In a simple word, optimization is an effective approach to achieve a “best” solution, where a single objective is maximized or minimized.

The technqiue used for optimization is Batch Gradient Descent.

####  Gradient Descent

Gradient descent (also known as steepest descent) is a first-order iterative optimization algorithm for locating a differentiable function's local minimum. The idea is to take repeated steps in the opposite direction of the function's gradient (or approximate gradient) at the current point, because this is the steepest descent. Stepping in the direction of the gradient, on the other hand, will result in a local maximum of that function; this procedure is known as gradient ascent. The technqiue used for optimization is Gradient Descent - Batch Gradient Descent.

#### Batch Gradient Descent

Batch gradient descent is a variation of the gradient descent algorithm that calculates the error for each example in the training dataset, but only updates the model after all training examples have been evaluated.

One cycle through the entire training dataset is called a training epoch. Therefore, it is often said that batch gradient descent performs model updates at the end of each training epoch.

#### Explanation on the Batch Gradient Descent (Iris-setosa species)

First of all, the dataframe is reshaped and assigned in to m and n. Secondly, the first 50 data from dataframe which are   Iris-sentosa species data. Two variables(sepal length and sepal width) of these 50 data are assigned into X and Y variable. Next, the hstack function is used to concatenate along the second axis. Hstack is a function in python which is stack arrays in sequence horizontally (column wise). 

Next, the np.random.seed(0) is used to make the random number predictable. The theta is equal to np.random.randn(1,2).Np.random.randn will return a sample (or samples) from the "standard normal" distribution. Then, the theta is printed out.

The iteration used to run is set to 10000. The np.zeros returns a new array of given shape and type, filled with zeros.It means that will run the iteratio for 10000 times. The learning rate defined is 0.001.The smaller learning rate the longer Gradient Descent converges, or may reach maximum iteration before reaching the optimum point. If learning rate is too big the algorithm may not converge to the optimal point (jump around) or even to diverge completely.
Thus, a learning rate with 0.001 is fine for conduct Gradient Descent Optimization to reach an optimum point. 

After that, the training started to train the data. The x-axis indicates the number of iterations while the y-axis indicates cost.  Gradient Descent will apply the formula to find a local minimum point.

```python
m,n=df.shape

#taking only first 50 columns of Iris-setosa species
X=df.SepalLengthCm.iloc[0:50].values.reshape(50,1)
Y=df.SepalWidthCm.iloc[0:50].values.reshape(50,1)

Xnew=np.hstack((np.ones((50,1)),X))

np.random.seed(0)
theta=np.random.randn(1,2)
print('Theta:',theta)

iters=10000
J=np.zeros(iters)
learning_rate=0.001

#training

for i in range(iters):
    J[i]=(1/(2*m))*np.sum((np.dot(Xnew,theta.T)-Y)**2)
    theta[0,0]-=(learning_rate/m)*np.sum(np.dot(Xnew,theta.T)-Y)
    theta[0,1]-=(learning_rate/m)*np.sum((np.dot(Xnew,theta.T)-Y)*X)
plt.plot(np.arange(iters),J)
plt.xlabel('Number of iterations',color='red')
plt.ylabel('Cost',color='red')
plt.title('Cost vs iterations',color='green')
plt.show()
```

Based on the graph above, the cost is decreasing from iteration to iteration and lastly reached a local minimum point with 0.0150.

#### Plot scatter graph on the Gradient Descent (Iris-setosa species)

```python
print('Updated theta after Gradient Descent:',theta)
plt.scatter(X,Y,marker='.')
plt.plot(X,np.dot(Xnew,theta.T))
plt.xlabel('Sepal Length(in cm)')
plt.ylabel('Sepal Width(in cm)')
plt.title('Sepal length vs width of Iris-Setosa Species',color='green')
plt.show()
```

Based on the scatter plot above, we can see that the theta value have been updated from [1.76405235 0.40015721] to [1.70887041 0.1788304]. This is because, after implementing batch gradient descent, it minimize the loss function.

#### Explanation on the Batch Gradient Descent (Iris-versicolor species)

First of all, the dataframe is reshaped and assigned in to a and b. Secondly, the first 50 data from dataframe which are Iris-versicolor species data. Two variables(sepal length and sepal width) of these 50 data are assigned into A and B variable. Next, the hstack function is used to concatenate along the second axis. Next, the np.random.seed(0) is used to make the random number predictable. The theta is equal to np.random.randn(1,2).Np.random.randn will return a sample (or samples) from the "standard normal" distribution. Then, the theta is printed out.

The iteration used to run is set to 10000. The np.zeros returns a new array of given shape and type, filled with zeros.It means that will run the iteratio for 10000 times. The learning rate defined is 0.001.The smaller learning rate the longer Gradient Descent converges, or may reach maximum iteration before reaching the optimum point. If learning rate is too big the algorithm may not converge to the optimal point (jump around) or even to diverge completely. Thus, a learning rate with 0.001 is fine for conduct Gradient Descent Optimization to reach an optimum point.

After that, the training started to train the data. The x-axis indicates the number of iterations while the y-axis indicates cost. Gradient Descent will apply the formula to find a local minimum point.

```python
a,b=df.shape
A=df.SepalLengthCm.iloc[51:100].values.reshape(49,1)
B=df.SepalWidthCm.iloc[51:100].values.reshape(49,1)

Xnew=np.hstack((np.ones((49,1)),A))

np.random.seed(0)
theta=np.random.randn(1,2)
print('Theta:',theta)

iters=10000
J=np.zeros(iters)
learning_rate=0.001

#training

for i in range(iters):
    J[i]=(1/(2*a))*np.sum((np.dot(Xnew,theta.T)-B)**2)
    theta[0,0]-=(learning_rate/a)*np.sum(np.dot(Xnew,theta.T)-B)
    theta[0,1]-=(learning_rate/a)*np.sum((np.dot(Xnew,theta.T)-B)*A)
plt.plot(np.arange(iters),J)
plt.xlabel('Number of iterations',color='red')
plt.ylabel('Cost',color='red')
plt.title('Cost vs iterations',color='green')
plt.show()
```

The learning rate defined is 0.001.The smaller learning rate the longer Gradient Descent converges, or may reach maximum iteration before reaching the optimum point. If learning rate is too big the algorithm may not converge to the optimal point (jump around) or even to diverge completely. Thus, a learning rate with 0.001 is fine for conduct Gradient Descent Optimization to reach an optimum point.

After that, the training started to train the data. The x-axis indicates the number of iterations while the y-axis indicates cost. Gradient Descent will apply the formula to find a local minimum point.

```python
c,d=df.shape

C=df.SepalLengthCm.iloc[101:150].values.reshape(49,1)
D=df.SepalWidthCm.iloc[101:150].values.reshape(49,1)

Xnew=np.hstack((np.ones((49,1)),C))

np.random.seed(0)
theta=np.random.randn(1,2)
print('Theta:',theta)

iters=10000
J=np.zeros(iters)
learning_rate=0.001

#training

for i in range(iters):
    J[i]=(1/(2*c))*np.sum((np.dot(Xnew,theta.T)-D)**2)
    theta[0,0]-=(learning_rate/c)*np.sum(np.dot(Xnew,theta.T)-D)
    theta[0,1]-=(learning_rate/c)*np.sum((np.dot(Xnew,theta.T)-D)*C)
plt.plot(np.arange(iters),J)
plt.xlabel('Number of iterations',color='red')
plt.ylabel('Cost',color='red')
plt.title('Cost vs iterations',color='green')
plt.show()
```

Based on the graph above, the cost is decreasing from iteration to iteration and lastly reached a local minimum point around 0.02.

Based on the scatter plot above, we can see that the theta value have been updated from [1.76405235 0.40015721] to [1.72251836 0.18922491]. This is because, after implementing batch gradient descent, it minimize the loss function.
